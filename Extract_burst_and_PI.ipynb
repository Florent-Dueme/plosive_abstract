```python
import pandas as pd
import os
```

# Extract data

```python
PATH_bil = r'C:\Users\Dell\Desktop\Extraction\Burst or not bilinguals'
```

```python
DataList = []

for file in os.listdir(PATH_bil):
    if ".csv" in file:
        with open(os.path.join(PATH_bil, file), "r") as f:
            next(f)
            for line in f:
                listed = line.replace("\n", "").split(";")
                DataList.append(listed)


Dataframe_bil = pd.DataFrame(DataList, columns=["File", "Plosion index", "VOT no release", "VOT release"])
```

```python
display(Dataframe_bil)
```

```python
PATH_mono = r'C:\Users\Dell\Desktop\Extraction\Burst or not monolinguals'
```

```python
DataList = []

for file in os.listdir(PATH_mono):
    if ".csv" in file:
        with open(os.path.join(PATH_mono, file), "r") as f:
            next(f)
            for line in f:
                listed = line.replace("\n", "").split(";")
                DataList.append(listed)


Dataframe_mono = pd.DataFrame(DataList, columns=["File", "Plosion index", "VOT no release", "VOT release"])
```
