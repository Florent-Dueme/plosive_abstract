```python
import pandas as pd
```

```python
all = pd.read_csv("frspbil_plosives_22_09_2021.csv", sep=";", encoding="ISO-8859-1")
burst = pd.read_csv("clean_PI_burst_23_01_2022.csv", sep=";", encoding="ISO-8859-1")
```

```python
all.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', "Problem", "Fricatived"], inplace=True)
burst.drop(columns=['Error'], inplace=True)
burst["VOT"] = burst["VOT"] * 1000
burst["Length"] = burst["Length"] * 1000
```

```python
all["Burst"] = ""
all.rename(columns={"Plosive_Index": "Plosion_index"}, inplace=True)
burst.rename(columns={"Plosion index": "Plosion_index"}, inplace=True)
```

```python
def get_bin(x):
    if x:
        return 1
    else:
        return 0

burst["Burst"] = burst["Burst"].apply(get_bin)
```
